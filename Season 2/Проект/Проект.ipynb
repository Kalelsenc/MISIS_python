Люди во многом ассоциировали себя с песнями и музыкой. Это может улучшить настроение, уменьшить боль и беспокойство и облегчить возможности для эмоционального выражения. Исследования показывают, что музыка может принести пользу нашему физическому и психическому здоровью во многих отношениях.

В последнее время было проведено множество исследований, чтобы понять песни и их популярность на основе определенных факторов. Такие сэмплы песен разбиваются на части, и их параметры записываются для составления таблиц. Прогнозирование популярности песни является основной целью.

Проект простой, но сложный, чтобы предсказать популярность песни на основе энергии, акустики, инструментальности, живости, танцевальности и т. д. Набор данных большой, и его сложность возникает из-за того, что он имеет сильную мультиколлинеарность.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import math

df1 = pd.read_csv('song_data.csv', delimiter=',')
df = df1.copy()
df
```

Разберём параметры нашего датасета

- song_name (Название песни) - говорит само за себя, но в нашем исследовании будет удален.
- song_popularity (Популярность песни) - измеряется от 0 до 100, показывает насколько популярна (100 очень популярная).
- song_duration_ms - длительность песни в мс.
- acousticness (Акустичность) - описывает, насколько акустической является песня. Оценка 1 означает, что песня, скорее всего, будет акустической.
- danceability (Танцевальность) - описывает, насколько трек подходит для танцев, основываясь на сочетании музыкальных элементов, включая темп, стабильность ритма, силу удара и общую регулярность. Значение 0 соответствует наименее танцевальному, а значение 1 наиболее танцевальному.
- energy (Энергичность) - представляет перцептивную меру интенсивности и активности. Как правило, энергичные треки кажутся быстрыми, громкими и шумными.
- instrumentalness (Инструментальность) - это значение представляет количество вокала в песне. Чем он ближе к 1, тем инструментальнее песня.
- liveness (Живость) - это значение описывает вероятность того, что песня была записана с живой аудиторией, значение выше 0,8 дает высокую вероятность того, что трек имеет живой звук.
- loudness (громкость) - количестов дБ относительно опорного уровня, превышающий опорный со знаком «плюс» (например, +3 дБ), а меньший опорного в децибелах со знаком «минус» (например, −6 дБ).
- speechiness (Речевость) - определяет наличие произнесенных слов в треке. Если речь песни выше 0,66, она, вероятно, состоит из произнесенных слов, оценка от 0,33 до 0,66 — это песня, которая может содержать как музыку, так и слова, а оценка ниже 0,33 означает, что в песне нет речи.
- tempo (Темп) - скорость исполнения музыкального произведения, измеряется в ударах в минуту.
- audio_valence (Валентность) - Показатель от 0,0 до 1,0, описывающий музыкальную позитивность, передаваемую треком. Треки с высокой валентностью звучат более позитивно (например, счастливые, веселые, эйфорические), а треки с низкой валентностью звучат более негативно (например, грустные, подавленные, злые).

```python
df = df.drop_duplicates(keep='first')
df
```

```python
features1 = continuous_features

for i in features1:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    df = df[df[i] <= (Q3+(1.5*IQR))]
    df = df[df[i] >= (Q1-(1.5*IQR))]
    df = df.reset_index(drop=True)
display(df.head())
```

## Теперь перейдём к кластеризации

Посмотрим на количество кластеров с помощью метода локтя

```python
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Сумма квадратов расстояний')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(data_transformed)
```

## Построение предсказательной модели

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

dataLR = df.drop(columns = ['song_name'])
target = 'song_popularity'
X = dataLR.drop([target],axis=1)
Y = dataLR[target]
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=100)
Train_X.reset_index(drop=True,inplace=True)

print('Изначальные данные  ---> ',X.shape,Y.shape,'\nДанные для тренировки  ---> ',Train_X.shape,Train_Y.shape,'\nДанные для теста   ---> ', Test_X.shape,'', Test_Y.shape)
```

```python
std = StandardScaler()

print('\033[1mСтандартизация на тренировочных данных'.center(120))
Train_X_std = std.fit_transform(Train_X)
Train_X_std = pd.DataFrame(Train_X_std, columns=X.columns)
display(Train_X_std.describe())

print('\n','\033[1mСтандартизация на тестовых данных'.center(120))
Test_X_std = std.transform(Test_X)
Test_X_std = pd.DataFrame(Test_X_std, columns=X.columns)
display(Test_X_std.describe())
```

Сделаем отображение данных

```python
rc=np.random.choice(Train_X_std.loc[:,Train_X_std.nunique()>=50].columns.values,3,replace=False)
def Evaluate(n, pred1,pred2):
    plt.figure(figsize=[15,6])
    for e,i in enumerate(rc):
        plt.subplot(2,3,e+1)
        plt.scatter(y=Train_Y, x=Train_X_std[i], label='Реальные данные')
        plt.scatter(y=pred1, x=Train_X_std[i], label='Предсказанные данные')
        plt.legend()
    plt.show()
```

Множественная линейная регрессия

```python
MLR = LinearRegression().fit(Train_X,Train_Y)
pred1 = MLR.predict(Train_X)
pred2 = MLR.predict(Test_X)

print('Коэффициенты регрессии равны ',MLR.coef_)
print('Константа модели ',MLR.intercept_)

Evaluate(0, pred1, pred2)
```

```python
print('Значение R2 для данной модели ',round(r2_score(Train_Y, pred1),20))
```

Модель полиномиальной регрессии

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(Train_X_std)
X_poly1 = poly_reg.fit_transform(Test_X_std)
PR = LinearRegression()
PR.fit(X_poly, Train_Y)

pred1 = PR.predict(X_poly)
pred2 = PR.predict(X_poly1)

print('Коэффициенты регрессии равны ',PR.coef_)
print('Константа модели ',PR.intercept_)

Evaluate(4, pred1, pred2)
```

```python
print('Значение R2 для данной модели ',round(r2_score(Train_Y, pred1),20))
```
